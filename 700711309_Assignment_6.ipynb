```python
import pandas as pd #importing pandas
import numpy as np #importing numpy
from sklearn.preprocessing import StandardScaler,normalize # importing standard scaler and normalize
from sklearn.decomposition import PCA #importing PCA
from sklearn.metrics import silhouette_score #Importing silhouette_score
import warnings # importing warinings
warnings.filterwarnings('ignore') # to ignore all the warnings
from matplotlib import pyplot as plt #importing matlab
```

```python
#question 2
data = pd.read_csv('CC GENERAL.csv') #reading the data using pandas
data.head() #display first five rows

```

```python
#2.1 Preprocess the data by removing the categorical column and filling the missing values
data.fillna(data.mean(),axis=0,inplace=True) #replacing null velues with mean
X = data.drop(columns=['CUST_ID','TENURE'])#considering x values and dropping required columns
X.head() #displaying first 5 x rows
```

```python
#2.2 Apply StandardScaler() and normalize() functions to scale and normalize raw input data.

scaler = StandardScaler() # creating a StandardScalar to scale the data
scaler.fit(X) # feeding the data to the scalar
X_scaled_array = scaler.transform(X) # scaling the data using Standard Scalar function
X_scaled = pd.DataFrame(X_scaled_array, columns = X.columns) # creating a new dataframe with scaled data
X_scaled.head()#displaying first 5 rows of scaled x values
```

```python
X_normal_array = normalize(X_scaled) #normalizing the scaled data
X_normal = pd.DataFrame(X_normal_array,columns = X.columns) #creating a df with column as X
X_normal.head() #displaying first 5 rows of x scaled.
```

```python
#2.3 Use PCA with K=2 to reduce the input dimensions to two features.


pca2 = PCA(n_components=2) # creating PCA with no of components is 2
principalComponents = pca2.fit_transform(X_normal) # giving the data to PCA
# Creating new dataframe from the result obtained from PCA
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
# principalDf.head()
final_data = pd.concat([principalDf, data[['TENURE']]], axis = 1) # concatinating the PCA values and target values
final_data.head() # displaying first 5 rows of finalcc
```

```python
X_pca = final_data.drop('TENURE',axis=1) # dropping 'TENURE' from finalcc and storing as X_pca
y_pca = final_data['TENURE'] # considering 'TENURE' and storing as y_pca
```

```python
#2.4 Apply Agglomerative Clustering with k=2,3,4 and 5 on reduced features and visualize
#result for each k value using scatter plot.

from sklearn.cluster import AgglomerativeClustering #importing Agglomerative clustering
k = [2,3,4,5] #considering k from 2 to 5
result = [] #creating a result array
for i in k: #creating a for loop
    cluster = AgglomerativeClustering(i).fit(X_pca,y_pca) #inputting x_pca and y_pca values
    y_pred = cluster.fit_predict(X_pca) #prdicting y values
    result.append(y_pred) #appening y_pred to result
```

```python
for i in result: #creating for loop
    plt.scatter(X_pca['principal component 1'],X_pca['principal component 2'],c = i) #creating scatter plot for PCA components
    plt.show() #display
```

```python
#2.5 Evaluate different variations using Silhouette Scores and Visualize results with a bar chart.

silhouette_scores = [] #new array as silhouette score
for i in result: #creating for loop
    score = silhouette_score(X_pca, i) #Inputting x_pca and i values
    silhouette_scores.append(score)#appening the score values
    print(f'Silhouette score for {len(np.unique(i))} is {score}' ) #printing the values for all K ie., from 2 to 5
```

```python
plt.bar([2,3,4,5],silhouette_scores) #plotting bar for silhouette_score
```
